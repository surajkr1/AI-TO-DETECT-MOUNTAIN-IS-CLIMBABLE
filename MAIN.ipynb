SIMPLE LOGIC FOR THIS AI.
MOUNTAINS WHICH ARE UNCLIMBED ARE ALREADY HIGHLIGHTED IN THIS DATASET IN THE COLUMN ASCENTS FIRST AND REMAINING ALL MOUNTAINS ARE ONCE OR MORE TIMES CLIMBED.
SO WE ARE GOING TO CHECK IF MOUNTAINS ARE NEVER CLIMBED THEN IT WILL PRINT UNCLIMBABLE AND IF MOUNTAINS ARE CLIMBED EVEN ONCE THEN IT WILL PRINT CLIMBABLE.

# AI

```python
while True:
    name = input("Enter Mountain Name: ")
    if name in list(data['Mountain name(s)']):
        d = data.loc[data['Mountain name(s)'] == name]
        x = list(d['ascents_first'])
        if 'unclimbed' in x :
            print("Unclimable")
        else:
            print("Climable")
    else:
        print("Name Of Mountain Provided is Wrong or not in the trained mountain list.")
```
